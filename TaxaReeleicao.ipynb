```python
file = 'votacao_candidato_munzona_2016.zip'

data = []
data_dict = {}
with zipfile.ZipFile(file) as z:
    for filename in z.namelist():
        if not os.path.isdir(filename):
            with z.open(filename) as f:
                #data.append(f.read().decode('Latin-1')) #('ISO-8859-1'))
                text = []
                for line in f:
                    text.append(line.decode('latin-1').strip().replace('"','').split(sep=';'))
                data.append(text)
                data_dict[filename[-6:-4]] = pd.DataFrame(text)
                    

#' '.join(data[0])
```
